```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
y = iris.target
x = iris.data
```

```python
pca = PCA(n_components = 2)
pca = pca.fit(x)
x_dr = pca.transform(x)
```

```python
x_dr[y == 0,0]
```

```python
iris.target_names[1]
```

```python
plt.figure()
plt.scatter(x_dr[y==0,0],x_dr[y==0,1],color="red",label = iris.target_names[0])
plt.scatter(x_dr[y==1,0],x_dr[y==1,1],color="black",label = iris.target_names[1])
plt.scatter(x_dr[y==2,0],x_dr[y==2,1],color="green",label = iris.target_names[2])
plt.legend()
plt.title("PCA of IRIS data")
plt.show()
```

```python
colors = ['red', 'black', 'orange']
for i in range(3):
    plt.scatter(x_dr[y==i,0],x_dr[y==i,1],c=colors[i],label = iris.target_names[i],alpha=.5)
plt.legend()
plt.title("PCA of IRIS data")
plt.show()    
```

```python
def plot_digits(data):
    fig,axes = plt.subplots(4,10,figsize=(10,4)
                            ,subplot_kw = {"xticks":[],"yticks":[]}
                            
                            )
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap="binary")
```

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import cross_val_score 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"D:\jupyterwork\caicai_class\digit recognizor.csv")
```

```python
x = data.iloc[:,1:]
y= data.iloc[:,0]
```

```python
x.shape
```

```python
pca_line = PCA().fit(x)
plt.figure(figsize=[20,5])
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance")
plt.show()
```

```python
score = []
for i in range(1,101,10):
    x_dr = PCA(n_components=i).fit_transform(x)
    once = cross_val_score(RFC(n_estimators=10, random_state=0),x_dr,y,cv=5).mean()
    score.append(once)
```

```python
plt.figure(figsize=[20,5])
plt.plot(range(1,101,10),score)
plt.show()
```

```python
score = []
for i in range(10,25):
    x_dr = PCA(n_components=i).fit_transform(x)
    once = cross_val_score(RFC(n_estimators=10, random_state=0),x_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(10,25),score)
plt.show()
```
